```python
import numpy as np
import pandas as pd
import json
from datetime import datetime
import pytz

data= pd.read_csv("H:\IPMC.TXT")
data.columns= ["unixtime","temp","rh","pnp3","pnp5","pn1","pn2p5","pn5","pn10","pm1","pm2p5","pm10"]
data['time'] = pd.to_datetime(data['unixtime'],unit='s')
data.head()
```

# Particle Mass Concentration

```python
plt.rcParams['figure.figsize'] = (10,5)
plt.style.use('seaborn')
plt.figure()

fig, ax = plt.subplots()
ax.plot(data['time'],data['pm2p5'], 'b-')

plt.legend(loc='best', fontsize = 17)
plt.title('$PM2.5$', fontsize=20)
plt.xlabel('$Time$', fontsize=17)
plt.ylabel('PM2.5', fontsize=17)
plt.ylim(0,50)


plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1.plot(data['time'],data['pm10'], 'b-')

plt.legend(loc='best', fontsize = 17)
plt.title('$PM10$', fontsize=20)
plt.xlabel('$Time$', fontsize=17)
plt.ylabel('PM10', fontsize=17)
plt.ylim(0,50)


plt.show()
```

# Particle Number concentration

```python
fig, ax2 = plt.subplots()
ax2.plot(data['time'],data['pn2p5'], 'b-')

plt.legend(loc='best', fontsize = 17)
plt.title('$PN2.5$', fontsize=20)
plt.xlabel('$Time$', fontsize=17)
plt.ylabel('PN2.5', fontsize=17)
plt.ylim(0,50)


plt.show()
```
